```python
import plotly.graph_objects as go
import numpy as np

# Create a figure
fig = go.Figure()

# Set the background color to dark
fig.update_layout(
    plot_bgcolor='black',
    paper_bgcolor='black',
)

# Set the x and y limits
fig.update_xaxes(range=[-500, 500])
fig.update_yaxes(range=[-500, 500])

# Set the grid lines to neon color and thin
fig.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lime')
fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lime')

# Set the grid lines to be visible only when hovering over the grid
fig.update_layout(
    hovermode='x unified',
    hoverlabel=dict(
        bgcolor='white',
        font_size=26,
        font_family='Rockwell',
        font_color='lime'  # Set the font color to neon color
    )
)

# Show the plot
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Create a figure
fig = go.Figure()

# Set the background color to dark
fig.update_layout(
    plot_bgcolor='black',
    paper_bgcolor='black',
)

# Set the x and y limits
fig.update_xaxes(range=[-500, 500])
fig.update_yaxes(range=[-500, 500])

# Set the grid lines to neon color and thin
fig.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lime')
fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lime')

# Set the hover template to display the coordinates with neon color and larger font size
fig.update_layout(
    hovermode='x unified',
    hoverlabel=dict(
        bgcolor='white',
        font_family='Rockwell',
        font=dict(
            size=26,
            color='lime'
        )
    )
)

# Add a scatter plot to display the coordinates on hover
fig.add_trace(go.Scatter(
    x=[0],
    y=[0],
    mode='markers',
    hoverinfo='x+y',
    hoverlabel=dict(
        font=dict(
            size=26,
            color='lime'
        )
    )
))

# Show the plot
fig.show()
```
